```python
import pandas as pd
import re
from collections import Counter

user = 'Khairykj'

# Filter the DataFrame based on the specified criteria
filtered_df = result_pol[(result_pol['Conversation ID'] ==result_pol['User ID']) & (result_pol['Username'] == user) & (result_pol['status'] == 'Main post')]

# Get the conversation ID values as a list
conversation_ids = filtered_df['Conversation ID'].tolist()

# Filter for rows where the conversation ID is in the list of conversation IDs and the status is "reply"
replies = result_pol[(result_pol["Conversation ID"].isin(conversation_ids)) & (result_pol["status"] == "reply")]

# Extract mentions from the "Text" column using regular expressions and remove the mentions to the user
mentions = replies["Text"].apply(lambda x: [mention for mention in re.findall(r"@(\w+)", x) if mention != user]).sum()

# Count the appearance of each username and polarity using the Counter class
mentions_count = Counter(mentions)
positive_count = Counter(replies[replies['sentiment_label'] == 'Positive']["Text"].apply(lambda x: [mention for mention in re.findall(r"@(\w+)", x) if mention != user]).sum())
negative_count = Counter(replies[replies['sentiment_label'] == 'Negative']["Text"].apply(lambda x: [mention for mention in re.findall(r"@(\w+)", x) if mention != user]).sum())

# Convert the Counter objects to DataFrames and sort by the count
mentions_df = pd.DataFrame.from_dict(mentions_count, orient="index", columns=["Total count"]).sort_values(by="Total count", ascending=False)
positive_df = pd.DataFrame.from_dict(positive_count, orient="index", columns=["Positive count"]).sort_index().astype(int)
negative_df = pd.DataFrame.from_dict(negative_count, orient="index", columns=["Negative count"]).sort_index().astype(int)

# Join the DataFrames
result_df = mentions_df.join([positive_df, negative_df], how='outer')

# Print the resulting DataFrame
print(result_df.head(10))
```

```python
import pandas as pd
from collections import Counter

users = ['chairman_GLC', 'Khairykj', 'MuhyiddinYassin', 'limlipeng', 'kuasasiswa', 'NajibRazak', 'n_izzah', 'anwaribrahim', 'SyedSaddiq']

for user in users:
    # Filter the DataFrame based on the specified criteria
    filtered_df = result_pol[(result_pol['Conversation ID'] == result_pol['User ID']) & (result_pol['Username'] == user) & (result_pol['status'] == 'Main post')]

    # Get the conversation ID values as a list
    conversation_ids = filtered_df['Conversation ID'].tolist()

    # Filter for rows where the conversation ID is in the list of conversation IDs and the status is "reply"
    replies = result_pol[(result_pol['Conversation ID'].isin(conversation_ids)) & (result_pol['status'] == 'reply')]

    # Count the appearance of each username and polarity using the Counter class
    username_counts = Counter(replies['Username'].str.cat(sep=',').split(','))
    positive_counts = Counter(replies[replies['sentiment_label'] == 'Positive']['Username'].str.cat(sep=',').split(','))
    negative_counts = Counter(replies[replies['sentiment_label'] == 'Negative']['Username'].str.cat(sep=',').split(','))

    # Remove the user's own username from the counter
    del username_counts[user]
    del positive_counts[user]
    del negative_counts[user]

    # Convert the Counter objects to DataFrames and sort by the count
    username_df = pd.DataFrame.from_dict(username_counts, orient='index', columns=['count']).sort_values(by='count', ascending=False)
    positive_df = pd.DataFrame.from_dict(positive_counts, orient='index', columns=['positive_replies']).sort_index().astype(int)
    negative_df = pd.DataFrame.from_dict(negative_counts, orient='index', columns=['negative_replies']).sort_index().astype(int)

    # Join the two DataFrames
    result_df = username_df.join([positive_df, negative_df], how='outer')

    # Print the resulting DataFrame
    print(f"Top 10 usernames mentioned in replies to {user}:")
    print(result_df.head(10))
    print("-------------------------------------------------------------------------------")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Define the list of texts
texts = davidwong27_rows['translated cleaned'].tolist()

# Create an instance of TfidfVectorizer
tfidf = TfidfVectorizer()

# Fit the vectorizer on the texts
tfidf.fit(texts)

# Transform the texts to a TF-IDF matrix
tfidf_matrix = tfidf.transform(texts)

# Get the feature names (i.e. the words)
feature_names = tfidf.get_feature_names()

# Get the TF-IDF scores for each word in each text
scores = tfidf_matrix.toarray()

# Sum the scores for each word across all texts
word_scores = dict(zip(feature_names, scores.sum(axis=0)))

# Sort the words by their scores
top_words = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)[:15]

# Print the top 15 words and their scores
for word, score in top_words:
    print(f"{word}: {score}")
```
