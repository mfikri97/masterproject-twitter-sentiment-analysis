```python
from deep_translator import GoogleTranslator

def translate_dataframe_column(df, source_col, target_col, source_lang, target_lang):
    """
    Translates the text in a DataFrame column from the source language to the target language
    using Google Translator and stores the translated text in another column.

    Parameters:
    df (pandas.DataFrame): The DataFrame containing the text to be translated.
    source_col (str): The name of the column containing the text to be translated.
    target_col (str): The name of the column to store the translated text.
    source_lang (str): The source language code (e.g. 'en' for English, 'fr' for French).
    target_lang (str): The target language code (e.g. 'en' for English, 'fr' for French).

    Returns:
    pandas.DataFrame: The modified DataFrame with the translated text stored in the target column.
    """
    translator = GoogleTranslator(source=source_lang, target=target_lang)
    df[target_col] = df[source_col].apply(lambda x: translator.translate(x))
    return df
```
