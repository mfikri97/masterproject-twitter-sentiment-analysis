```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Create a new instance of the VADER analyzer
analyzer = SentimentIntensityAnalyzer()

# Convert the 'Translated Cleaned Text' column to string type
df_merged['translated cleaned'] = df_merged['translated cleaned'].astype(str)

# Apply the polarity_scores() method to calculate sentiment scores for each text
df_merged['sentiment Scores'] = df_merged['translated cleaned'].apply(lambda text: analyzer.polarity_scores(text))

# Extract the compound score from the sentiment scores
df_merged['sentiment'] = df_merged['sentiment Scores'].apply(lambda scores: scores['compound'])
```

```python
sample = 'delicious'


# Apply the polarity_scores() method to calculate sentiment scores for the text
sentiment_scores = analyzer.polarity_scores(sample)

# Extract the compound score from the sentiment scores
sentiment = sentiment_scores['compound']

# Print the sentiment score
print("Sentiment score for the text: ", sentiment)
```

```python
# Import the SentimentIntensityAnalyzer class
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Create an instance of the SentimentIntensityAnalyzer class
analyzer = SentimentIntensityAnalyzer()

# Split the sample text into a list of words
words = sample.split()

# Create an empty list to store the polarity scores of each word
polarity_scores = []

# Loop through each word in the list of words
for word in words:
    # Apply the polarity_scores() method to calculate sentiment scores for the word
    sentiment_scores = analyzer.polarity_scores(word)
    # Extract the compound score from the sentiment scores
    sentiment = sentiment_scores['compound']
    # Append the compound score to the list of polarity scores
    polarity_scores.append(sentiment)

# Print the table of words and their corresponding polarity scores
print("Word\tPolarity Score")
for i in range(len(words)):
    print(words[i], "\t", polarity_scores[i])
```

```python
# Define a function to label the sentiment
def label_sentiment(row):
    if row['sentiment'] > 0.05:
        return 'Positive'
    elif row['sentiment'] < -0.05:
        return 'Negative'
    else:
        return 'Neutral'


# Apply the function to each row and create a new column
df_merged['sentiment_label'] = df_merged.apply(label_sentiment, axis=1)


df_merged
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize the VADER polarity analyzer
sid = SentimentIntensityAnalyzer()

# Analyze text as a whole
text = "I have trust issue problem"
whole_analysis = sid.polarity_scores(text)
print("Whole Text Analysis:")
print(whole_analysis)

# Analyze text word by word
words = text.split()
word_by_word_analysis = []

for word in words:
    word_analysis = sid.polarity_scores(word)
    word_by_word_analysis.append((word, word_analysis))

print("\nWord-by-Word Analysis:")
for word, analysis in word_by_word_analysis:
    print(f"Word: {word}\tAnalysis: {analysis}")
```
